## <font color=904848 size=5>2. 需求分析

<font size=4>&emsp;&emsp;用户消费行为特征一般分为三个部分：用户消费特征、用户分层、用户质量。
    
<font size=4>&emsp;&emsp;第一个维度：用户消费特征。主要包括三个部分。第一，用户整体消费分析，即分析所有客户他的一个行为习惯；第二，用户个体消费分析，即按照用户的订单ID或用户ID分析他/她的行为习惯；第三，用户消费周期，包含购买周期和生命周期两个部分，用户购买周期即用户再次购买商品的时间间隔，而用户生命周期是指用户从首次消费到完全离开商品的一个过程，通过这两个指标可以推断用户的存活和流失周期。
    
<font size=4>&emsp;&emsp;第二个维度：用户分层。在这一版块，我们需要对用户的价值和活跃度两部分进行分析。在用户价值分析中，需要利用当今企业较为广泛应用的RMF模型来进行相应的计算与分析；而用户活跃度，我们一般从新老用户、活跃用户、不活跃用户、回流用户与流失用户这几个层面展开分析。
  
<font size=4>&emsp;&emsp;第三个维度：用户质量。主要针对如何提高用户质量进行展开。包括用户复购率、回购率、留存率、消费频率、客户贡献率等，通过这些数据，我们可以直观清晰地看出平台目前的用户质量，并根据当前的状况对现有客户进行针对性的管理与维护，并做出最合适的调整，改变现有的经营方式，完善营销策略，从而达到增加营收的目的。

## <font color=904848 size=5>3. 用户分层

### &emsp;3.1 构建RFM模型

<font size=4>&emsp;&emsp;RFM模型是衡量客户价值和客户创利能力的重要工具和手段。在RFM模式中，R表示客户购买的时间有多远，F表示客户在时间内购买的次数,M表示客户在时间内购买的金额.该模型通过一个客户的近期购买行为、购买的总体频率以及花了多少钱三项指标来描述开客户的价值状况。RFM模型较为动态地显示了一个客户的全部轮廓，这对个性化的沟通和服务提供了依据。同时如果与该客户打交道的时间足够长，也能够较为精确地判断该客户的长期价值，通过改善三项指标的状况，从而为更多的营销决策提供支持。

<font size=4>&emsp;&emsp;为了进行精细化运营，当今企业一般会通过构建RFM模型来对用户价值指数进行计算。因此在这一阶段，我们也将为该健身平台构建RFM模型，以此衡量历史到当前用户贡献的收益，并作出相应的决策。  
    
<font size=3>·&emsp;<b>R</b> (Recency)：最近一次消费；  
<font size=3>·&emsp;<b>F</b> (Frequency)：消费频率；  
<font size=3>·&emsp;<b>M</b> (Monetary)：消费金额。
    
<font size=3>·&emsp;R值越大，最近一次消费发生的日期越久远，反之则日期越近；F值越大，用户的消费频率越高，反之则越低；M值越大，用户消费金额越高，用户价值越高，反之则用户价值越低。

<font size=4>&emsp;&emsp;通过构建RFM模型后分类的用户可分为以下8种：
![jupyter](rfm_2.png)

<font size=4>&emsp;&emsp;用思维导图形式呈现更为直观：
![jupyter](rfm_1.png)

```python
#构建RFM模型
rfm = df.pivot_table(index='user_id',
                    values=['order_dt','order_products','order_amount'],
                    aggfunc={'order_dt':'max',
                            'order_products':'sum',
                            'order_amount':'sum'})
rfm.head()
```

```python
#添加R、F、M列
rfm['R'] = round(-(rfm['order_dt']-rfm['order_dt'].max())/np.timedelta64(1,'D'),0)
rfm.rename(columns={'order_products':'F','order_amount':'M'},inplace=True)
rfm.head()
```

```python
#用户分类
def rfm_func(x):
    level = x.apply(lambda x:'1' if x>=1 else '0')
    label = level['R'] + level['F'] + level['M']
    d={'000':'一般挽留客户',
       '001':'重要挽留客户',
       '010':'一般保持客户',
       '011':'重要保持客户',
       '100':'一般发展客户',
       '101':'重要发展客户',
       '110':'一般价值客户',
       '111':'重要价值客户',
      }
    result = d[label]
    return result

rfm['label'] = rfm[['R','F','M']].apply(lambda x:x-x.mean()).apply(rfm_func,axis=1)
rfm.head()
```

```python
#按照分类绘制散点图
for label,grouped in rfm.groupby('label'):
    x = grouped['F']
    y = grouped['R']
    plt.scatter(x,y,label=label)
plt.legend()
plt.title('用户分层散点图')
plt.xlabel('F')
plt.ylabel('R')
```

```python
#按照分类绘制饼图
rfm_count = rfm.value_counts('label')
rfm_count.plot(kind='pie',title='各类分层用户占比图',fontsize=10,explode=(0.03,0.03,0.05,0.05,0.1,0.2,0.3),shadow=True,startangle=60,autopct='%1.1f%%')
```

**<font size=3>观察用户分类的可视化：</font>**  

<font size=3>·&emsp;从图中可以看出一般发展用户的占比超过了全部用户的一半，二者类用户的质量相对不高，后续应当尽量减少这类用户的占比，并用相应的方法，提高几类重要客户的占比，并且提高服务质量，对于现有的高质量用户进行维系。

<font size=4><b>针对这8类用户，一般有以下不同的决策：</b>  
    
<font size=3>&emsp;<b>·&emsp;重要价值客户</b>：最优质的客户，应提高客户满意度，提供更多定制化服务；  
    
<font size=3>&emsp;<b>·&emsp;重要保持客户</b>：近期消费间隔较长的客户，可以通过活动提高购买频率；  
    
<font size=3>&emsp;<b>·&emsp;重要发展客户</b>：近期无消费的重要客户，应当主动触达防止客户流失；  
    
<font size=3>&emsp;<b>·&emsp;重要挽留客户</b>：是潜在价值客户，应主动了解客户需求，尽可能挽留；  
    
<font size=3>&emsp;<b>·&emsp;一般价值客户</b>：忠诚客户但消费金额较低，应引导其提高消费频率；   
    
<font size=3>&emsp;<b>·&emsp;一般保持客户</b>：属于新客户，应利用消费券机制提升复购率；  
    
<font size=3>&emsp;<b>·&emsp;一般发展客户</b>：可以减少对这一类客户的关注；   
    
<font size=3>&emsp;<b>·&emsp;一般挽留客户</b>：流失客户，可以通过短信推广尝试召回；     
    

### &emsp;3.2 分层用户

<font size=4>&emsp;&emsp;用户分层解释：  
    
<font size=3>&emsp;·&emsp;<b>新老用户</b>(new)：第一次消费的用户为新用户，反之则为老用户；  
<font size=3>&emsp;·&emsp;<b>活跃用户</b>(active)：在某一时间窗口内有过消费行为的用户；   
<font size=3>&emsp;·&emsp;<b>不活跃用户</b>(unactive)：在某一时间窗口内没有消费行为的用户；     
<font size=3>&emsp;·&emsp;<b>回流用户</b>(return)：即回头客，其余为流失用户。 
    
<font size=3>&emsp;&emsp;&emsp;用户回流动作可以分为自主回流与人工回流2种，自主回流指用户自发的回流，而人工回流是由人为参与导致的。
    

```python
pivoted_counts = df.pivot_table(
                    index = 'user_id',
                    columns = 'month',
                    values = 'order_dt',
                    aggfunc = 'count'
                ).fillna(0)
df_purchase = pivoted_counts.applymap(lambda x:1 if x>0 else 0)
df_purchase.head()
```

```python
def active_status(data):
    status = []
    for i in range(12):
        if data[i] == 0:
            if len(status) == 0:
                status.append('unreg')  #unreg表示未注册用户
            else:
                if status[i-1] == 'unreg':
                    status.append('unreg')
                else:
                    status.append('unactive')
            pass
        else:
            if len(status) == 0:
                status.append('new')
            else:
                if status[i-1] == 'unactive':
                    status.append('return')
                elif status[i-1] == 'unreg':
                    status.append('new')
                else:
                    status.append('active')
    return pd.Series(status,df_purchase.columns)

purchase_states = df_purchase.apply(active_status,axis=1)
purchase_states.head()
```

```python
purchase_states_ct = purchase_states.replace('unreg',np.NaN).apply(lambda x:pd.value_counts(x))
purchase_states_ct.T.fillna(0).plot.area()
plt.title('用户分层面积图')
plt.ylabel('用户数量')
```

**<font size=3>观察用户分层面积图：</font>**  

<font size=3>·&emsp;从图中可以看出活跃用户的数量从始至终都较为平稳，说明该健身平台的用户中有一部分是稳定的忠实用户；   
<font size=3>·&emsp;新用户数量6至10月有明显上升，而在之后趋于平稳；   
<font size=3>·&emsp;回流用户作为最核心的用户，总体较少；   
<font size=3>·&emsp;不活跃用户在前4个月中数量不多，后续随着月份的推移，逐渐增多，占比也越来越大。

```python
#用户占比
rate = purchase_states_ct.T.fillna(0).apply(lambda x:x/x.sum(),axis=1)
plt.plot(rate['return'],label='return')
plt.plot(rate['active'],label='active')
plt.legend()
plt.title('活跃用户与回流用户占比图')
plt.xlabel('Month')
plt.ylabel('百分比%')
```

**<font size=3>观察活跃用户与回流用户占比图：</font>**  

<font size=3>·&emsp;回流用户：总体保持平稳状态，占比不超过5%，并非理想状态；  
<font size=3>·&emsp;活跃用户：19年4月至7月期间占比较高，在6月达到峰值，从6月至10月逐步下降，之后趋于平稳。
<font size=3>·&emsp;结合两者的占比，可以看出，活跃用户占比一直大于回流用户的占比，平台的运营存在问题，应当及时进行调整，在改进商品及服务的基础上加强营销推广策略，吸引更多回流用户，确保平台的正常发展。

## <font color=C08890 size=6>七、复购率和回购率分析

## <font color=904848 size=5>1. 复购率分析

<font size=4>&emsp;&emsp;<b>复购率的定义</b>：
    
<font size=4>&emsp;&emsp;在某时间窗口内重复消费用户（消费两次及以上的用户）在总消费用户中占比。  

<font size=3>&emsp;<b>&emsp;·&emsp;重复消费用户</b>：一个用户一天产生多笔付款交易，也算作重复消费用户。即按天非去重法，是目前b2c网站统计数据常用计算方法，相对计算出来的重复购买率要高于按天去重所得结果。


```python
purchase_r = pivoted_counts.applymap(lambda x:1 if x>1 else np.NaN  if x==0 else 0)
(purchase_r.sum()/purchase_r.count()).plot(figsize=(12,6))
plt.title('用户复购率趋势图')
plt.xlabel('月份')
plt.ylabel('复购率')
```

**<font size=3>观察用户复购率趋势图：</font>**  

<font size=3>·&emsp;复购率从4月开始快速下降，直至8月开始有缓慢回升。分析原因可能是因为7/8月份有许多新用户仅消费一次造成的。

```python
a,b=plt.subplots(figsize=(10,6))
b.plot(purchase_r.count())
b.plot(purchase_r.sum())
legends=['消费人数','二次消费以上用户人数']
b.legend(legends)
plt.title('每月消费和二次消费以上用户人数')
plt.xlabel('时间(月)')
plt.ylabel('用户数')
```

```python
(purchase_r.sum()/purchase_r.count()).mean()
```

**&emsp;<font size=3>平均用户复购率为71.38%</font>**  

## <font color=904848 size=5>2. 回购率分析

<font size=4>&emsp;&emsp;<b>回购率的定义</b>：
    
<font size=4>&emsp;&emsp;某一个时间窗口内有过消费行为，在下一个时间窗口仍旧消费的用户占比。

```python
def purchase_back(data):
    status = []
    for i in range(11):
        if data[i] == 1:
            if data[i+1] == 1:
                status.append(1)
            elif data[i+1] == 0:
                status.append(0)
        else:
            status.append(np.NaN)
    status.append(np.NaN)
    return pd.Series(status,df_purchase.columns)
            
purchase_b = df_purchase.apply(purchase_back,axis=1)
purchase_b.head()
```

```python
(purchase_b.sum()/purchase_b.count()).mean()
```

**&emsp;<font size=3>平均用户回购率为54.06%</font>**  

```python
#可视化
plt.figure(figsize=(10,8))

plt.subplot(211)
(purchase_r.sum()/purchase_r.count()).plot(label='复购率')
(purchase_b.sum()/purchase_b.count()).plot(label='回购率')
plt.legend()
plt.xlabel('月份')
plt.ylabel('百分比%')
plt.title('用户复购率和回购率对比图')

plt.subplot(212)
plt.plot(purchase_b.sum(),label='回购人数')
plt.plot(purchase_b.count(),label='购物总人数')
plt.xlabel('月份')
plt.ylabel('百分比%')
plt.legend()
```

**<font size=3>观察回购率图：</font>**  

<font size=3>·&emsp;相比于复购率，回购率的波动更大；   
<font size=3>·&emsp;回购率在7-9月到达最低谷，在之后有所回升；  
<font size=3>·&emsp;总体来说，每月购物总人数大于回购人数，且在6月至10月期间，两者人数相差较大，猜测是因为这几个月增加了许多仅消费一次的新用户造成的。

## <font color=904848 size=5>2. 分析存在的问题：

<font size=4>&emsp;&emsp;前面分析中用到的RFM模型存在着一定的局限性：对于模型中的F值，如果某用户是忠诚的活动用户，啥时候有活动就啥时候买东西；或者用户固定每个月购买理财产品，这些行为都会导致F值可能是随机产生的，或是人为操控的，不太具有代表性。对于模型中的M值，用户过去消费高并不一定代表未来消费高，这其中可能存在着一些变数；或者用户趁开业活动薅羊毛囤货，都可能带来较高的M值，但这样的较高值会对用户分类带来一定的误差影响。
